# Load dataset 

```python
import pandas as pd
import numpy as np
# Data1 = Excel file with sensed data from the Firebase database
data1 = pd.read_excel('/content/drive/MyDrive/Datos ultimas 3 sem hasta 21 de enero.xlsx')
data1
```

# Filters creation


## Clean data null or less than zero

```python
#First filter: removes rows with null data
data1 = data1.dropna() 
data1 
```

```python
#Second filter: removes rows with data less than or equal to zero for TDS 
#Filter creation
neg0tds = data1['Value.TDS'] > 0 
#Filter aplication
data1 = data1[neg0tds] 
data1
```

```python
#Third filter: removes rows with data less than or equal to zero for water temperature
#Filter creation
neg0temp = data1['Value.TEMP'] > 0 
#Filter aplication
data1 = data1[neg0temp] 
data1
```

## Date and time filters

```python
# Date and time division
data = data1['Value.fecha'].str.split(' ', expand = True) 
data1[['Date', 'Hour']] = data
#Hour, second and minute divisions
data1[['Horas', 'Minutos', 'Segundos']] = data1['Hour'].str.split(':', expand = True) 
data1
```

```python
#Creation of the filter for data that were not taken in the change of cycle
hora0 = data1['Minutos'].astype(int) == 00 
#Filter aplication
data1 = data1[hora0] 
data1
```

## Arithmetic mean and standard deviation filter

```python
#Calculate the standard deviation of water temperature
detemp = data1['Value.TEMP'].std() 
print (detemp)
#Calculate arithmetic mean of water temperature
promtemp = data1['Value.TEMP'].mean() 
print (promtemp)
#Standard deviation filter considering the arithmetical mean of water temperature
data1 = data1[(data1['Value.TEMP'] < (promtemp + (2.5*detemp))) & (data1['Value.TEMP'] > (promtemp - (2.5*detemp)))] 
print(data1)
```

```python
#Calculate the standard deviation of TDS
detds = data1['Value.TDS'].std() 
print (detds)
#Calculate arithmetic mean of TDS
promtds = data1['Value.TDS'].mean() 
print (promtds)
#Standard deviation filter considering the arithmetical mean of water temperature
data1 = data1[(data1['Value.TDS'] < (promtds + (2.5*detds))) & (data1['Value.TDS'] > (promtds - (2.5*detds)))] 
print(data1)
```

 ## Export filtered dataset

```python
# Selection only of the data columns that require to be part of the dataset
datafinal=pd.DataFrame() 
datafinal['Value.fecha'] = data1['Value.fecha']
datafinal['Value.TDS'] = data1['Value.TDS']
datafinal['Value.TEMP'] = data1['Value.TEMP']
datafinal['Date'] = data1['Date']
datafinal['Horas'] = data1['Horas']
datafinal['Minutos'] = data1['Minutos']
# Name of the excel file in which the filtered dataset will be stored
file_name = 'Limpia_RN.xlsx' 
# Convert file to excel and export
datafinal.to_excel(file_name) 
print('Dataset has been exported')
```
